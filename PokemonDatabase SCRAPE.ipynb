### My first goal was to just find the information I wanted Scraped put into a DataFrame

```python
response = requests.get('https://pokemondb.net/pokedex/pikachu')
soup = BeautifulSoup(response.content, 'lxml')
table = soup.find('table', {'class':'vitals-table'})
pokemon = {}
pokemon['#'] = table.find('tbody').find_all('tr')[0].text.split('\n')[2].strip()
pokemon['species'] = table.find('tbody').find_all('tr')[2].text.split('\n')[2]
all_poke.append(pokemon)
pd.DataFrame(all_poke)
```

### Created a loop that went to each of the 721 Pokemon in question and obtained their Number and Species and put them into a DataFrame.

```python
all_poke_loop = []

for poke in pokemon_list:
    response = requests.get('https://pokemondb.net/pokedex/{0}'.format(poke))
    soup = BeautifulSoup(response.content, 'lxml')
    table = soup.find('table', {'class':'vitals-table'})
    pokemon = {}
    pokemon['#'] = table.find('tbody').find_all('tr')[0].text.split('\n')[2].strip()
    pokemon['species'] = table.find('tbody').find_all('tr')[2].text.split('\n')[2]
    all_poke_loop.append(pokemon)
    all_poke = pd.DataFrame(all_poke_loop)
```
